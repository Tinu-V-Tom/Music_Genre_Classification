#Doing all the preprocessing steps in test data too

```python
# Replace question marks only in the 'mode 'and 'tempo'  columns
test[['mode', 'tempo']] = test[['mode', 'tempo']].replace('?', 'NaN')
test['tempo']=test['tempo'].astype(float)

freq=test.select_dtypes(include=['float'])
for i in freq:
    test[i]= test[i].fillna(test[i].median())

#filling missing values in col 'Embarked' with mode since its categorical
test['voice_gender']=test['voice_gender'].fillna(test['voice_gender'].mode()[0])

# Replace NaN values with the mode of the 'mode' column
mode = test['mode'].mode()[0]
test['mode'] = test['mode'].fillna(mode)

print(test['mode'].mode())
test['mode'] = test['mode'].str.replace('NaN','Major')
#filling missing values in col 'Embarked' with mode since its categorical
test['musician_category']=test['musician_category'].fillna(test['musician_category'].mode()[0])

test.drop(['instance_id','track_name'],axis=1,inplace=True)
```

```python
X=train.drop('music_genre',axis=1)
y=train['music_genre']
y.value_counts()
```

#Encoding

```python
X.head()
```

Here for encoding, we go with one-hot encoding since we have the nominal categorical values

```python
#one hot encoding with drop_first=True for avoiding multicollinearity
x=pd.get_dummies(X,drop_first=True)
```

```python
x.head()
```

```python

x.dtypes
```

Here after OHE the newly created columns are of bool data type. I just need to convert that into o and 1s

```python
for i in x.columns:
  if x[i].dtype=='bool':
    x[i]=x[i].astype(int)
```

```python
x.dtypes
```

```python
x.head()
```

```python
# Perform 80 : 20 Train-Test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=123,stratify=y)
#here stratify is used because from EDA we identified class imbalance is present
```

```python
# Observe the shapes
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

# Model Building and Evaluating

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#define logistic regression model
lr=LogisticRegression()
#training the model with train data
lr_model=lr.fit(x_train,y_train)
#predictions are made
y_pred=lr_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score, precision_score, recall_score, f1_score,classification_report
#Evaluating the model performance
print('Accuracy is: ',accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('f1 score is : ',f1_score(y_test,y_pred,average='macro'))
```

#Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
#define decision tree classification model
dt=DecisionTreeClassifier(random_state=123)
#training the model with train data
dt_model=dt.fit(x_train,y_train)
#predictions are made
y_pred=dt_model.predict(x_test)
y_pred
```

```python
# Evaluating model performace
print('Accuracy is: ',accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('f1 score is : ',f1_score(y_test,y_pred,average='macro'))
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
#define random forest classification model
rf=RandomForestClassifier(random_state=123)
#training the model with train data
rf_model=rf.fit(x_train,y_train)
#predictions are made
y_pred=rf_model.predict(x_test)
y_pred
```

```python
#evaluating the model
print('Accuracy is: ',accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('f1 score is : ',f1_score(y_test,y_pred,average='macro'))
```

#Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
# Define the gradient boosting classifier model
gb_clf = GradientBoostingClassifier(random_state=42)

# Train the model on the training data
gb_clf.fit(x_train, y_train)

# Make predictions on the test data
y_pred = gb_clf.predict(x_test)

```

```python
#evaluating the model
print('Accuracy is: ',accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('f1 score is : ',f1_score(y_test,y_pred,average='macro'))
```

#Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.05],
    'max_depth': [3, 4, 5],
    'subsample': [0.8, 0.9, 1.0],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
```

n_estimators: The number of boosting stages(trees) to build.

learning rate: The learning rates shrink the contribution of each tree by this factor.There is a trade-off between this and n_estimators.

max_depth: The maximum depth of each tree. it limits the no: of nodes in each tree.

subsample: Fraction of samples to be used for fitting each base learner.

min_samples_split: Minimum no: of samples to split internal nodes.

min_samples_leaf: Minimum no: of samples to be at leaf node.

```python

# Initialize the GradientBoostingClassifier
gbc = GradientBoostingClassifier()

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=gbc, param_grid=param_grid, scoring='f1', cv=5, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(x_train, y_train)

# Get the best estimator
best_gbc = grid_search.best_estimator_

# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)

# Make predictions on the test set
y_pred = best_gbc.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')  # Use 'weighted' for multi-class

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"F1 Score: {f1 * 100:.2f}%")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```
